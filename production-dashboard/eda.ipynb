```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_csv(r"dataset\manufacturing_dataset.csv")
```

```python
df.head(5)
```

```python
df['time_stamp'] = pd.to_datetime(df['time_stamp']).dt.date
```

```python
df = df[['time_stamp','operation_mode','defect_rate','speed']]
df
```

```python
df['up_time'] = df['operation_mode'].apply(lambda r: 1) # in minute
df
```

```python
df = df.groupby(['time_stamp','operation_mode']).agg({
    'defect_rate':'mean',
    'speed':'mean',
    'up_time':'sum'
}).reset_index()
df
```
